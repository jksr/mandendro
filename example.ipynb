```python
import scipy.cluster.hierarchy as sch
import numpy as np
```

```python
dists = np.array([662., 877., 255., 412., 996., 295., 468., 268.,
                  400., 754., 564., 138., 219., 869., 669.])
Z = sch.linkage(dists, 'complete')
labels = [f'c{x}' for x in range(len(Z)+1)]
```

```python
mdendro = ManDendro(Z,labels)
mdendro.plot_dendro()
```
